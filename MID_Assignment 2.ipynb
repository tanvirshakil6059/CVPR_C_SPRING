```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```python
#PReLU
def Leaky_ReLU(x):
    if x>0:
        return x
    else: 
        return 0.1*x

y = []
for i in x:
     y.append(Leaky_ReLU(i))
y = list(map(Leaky_ReLU, x))
plot_graph(y,"ReLU(X)")
```

```python
def elu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

y = elu(x)
plot_graph(y,"ELU(X)")
```

```python
def selu(x, alpha=1.67326324, scale=1.05070098):
    if x > 0:
        return scale * x
    else:
        return scale * alpha * (np.exp(x) - 1)
    
y = [selu(xi) for xi in x]
plot_graph(y, "SELU(X)")
```
